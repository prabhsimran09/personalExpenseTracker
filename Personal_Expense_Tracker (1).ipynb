```python
##### expense will be a list of dictionaries -> dictionary structure is defined as below
# dict1 = { "name":"", "date":[], "category":[], "amount":[], "description":[], "budget":[]}
import datetime
import csv

# create expense list
expenses=[]
# create expense list
past_expenses=[]

def add_expense(name) -> "Takes user_name, date_of_expense, category_of_expense, amount_of_expense, description_of_expense as input":
    '''This method enables user to add an expense'''

    
    print("***********************************************************************")
    print("############################ ADD EXPENSE ##############################") 
    print("***********************************************************************")
    print()
    print()
    print()
    print("Hi ",name," !! Please enter expense details below") 
    print()
    # create new expense
    expense={}
    expense["name"]=name
    

    # input date -> add exception handling for date format after tomorrow
    while(True):
        input_date = input("Enter date of expense (YYYY-MM-DD) : ") 
        print()
        if len(input_date) !=10 : 
            print()
            print("Invalid Date Format !! Please try again")
            print()
        else:
            try:
                date = datetime.date.fromisoformat(input_date)
                break
            except ValueError:
                print()
                print("Invalid isoformat string:",input_date)
                print()
    expense["date"]=date

    # input category
    print("Select category from the list given below")
    print()
    print("1-->Food")
    print("2-->Travel")
    print("3-->Entertainment")
    print("4-->Utility Bills")
    print("5-->Home Shopping")
    print("6-->Clothing")
    print("7-->Medical Expenses")
    print("8-->Rent")
    print("9-->Miscellaneous")
    print()
    while(True):
        input_category=int(input("Your category choice : "))
        if input_category == 1 :
            category ="FOOD"
            break;
        elif input_category == 2 :
            category="TRAVEL"
            break;
        elif input_category == 3 :
            category="ENTERTAINMENT"
            break;
        elif input_category == 4:
            category="BILLS"
            break;
        elif input_category == 5 :
            category="Shopping"
            break;
        elif input_category == 6 :
            category="Clothing"
            break;
        elif input_category == 7 :
            category="Medical Expenses"
            break;
        elif input_category == 8 :
            category="Rent"
            break;
        elif input_category == 9 :
            category="Miscellaneous"
            break;
        else:
            print()
            print("Invalid Category !! Please Try Again")
    expense["category"]=category

    # input amount
    print()
    amount = float(input("Enter the expense amount : ")) 
    expense["amount"]=amount

    # input description
    while(True):
        print()
        description = input("Enter a breif description ( max 30 words ): ")
        print()
        print()
        if len(description)>150:
            print()
            print("Oops, word limit exceeded!! Please Try again")
        elif len(description) == 0:
            print()
            print("Description can't be empty")
        else:
            break
    expense["description"]=description

    expenses.append(expense)

def view_expense(name):

    # iterate on list of expenses to 
    print("***********************************************************************")
    print("############################ VIEW EXPENSE #############################") 
    print("***********************************************************************")
    print()
    print()
    print()
    print("Hi ",name," !! Given below is a list of all expenses") 
    print()

    #  create a table of expenses
    for x in expenses:
        for item in x.items():
            print(item)
        print()
        print("***********************************************************************")
        print()
    
    print()
    print()


def total_expenses(name):
    month = datetime.datetime.now().month
    sum=0.0
    for x in expenses:
        if x["name"] == name and x["date"].month == month:
            sum+=x["amount"]
    for x in past_expenses:
        if x["name"] == name and x["date"].month == month:
            sum+=x["amount"]
    return sum

def track_budget(name):
    print("***********************************************************************")
    print("############################ TRACK BUDGET #############################") 
    print("***********************************************************************")
    print()
    print()
    print()
    print("Hi ",name,", Add budget details below !!")
    budget=float(input("Please enter the amount you want to budget for the month : "))
    total_expense=total_expenses(name)
    if total_expense == 0:
        print("No expense recorded for this month")
    if total_expense>budget :
        print("You have exceeded your budget!")
    else:
        print("You have ", budget-total_expense," left for the month")
    print()
    print()

def save_expense(name):
    print("***********************************************************************")
    print("############################# SAVE EXPENSES ###########################")
    print("***********************************************************************")
    print()
    print()
    print("Hi ",name,"Welcome to save Menu")
    print()
    while(True):
        print("Please select your choice : ")
        print() 
        print() 
        print("1-->Overwrite Saved Expenses")
        print() 
        print("2-->Keep previous expenses and update new ones")
        print() 
        print()
        choice=int(input("Enter your choice"))
        if choice == 1:
            file = open("data.csv","w")
            writer = csv.writer(file)
            writer.writerow(["NAME","DATE","CATEGORY","AMOUNT","DESCRIPTION"])
            for expense in expenses:
                writer.writerow([expense["name"],expense["date"],expense["category"],expense["amount"],expense["description"]])
            break
        elif choice==2:
            file=open("data.csv","a")
            writer = csv.writer(file)
            for expense in expenses:
                writer.writerow([expense["name"],expense["date"],expense["category"],expense["amount"],expense["description"]])
            break
        else:
            print("Invalid choice!!!!! Please Try Again")
    
def load_expense():
    file = open("data.csv","r")
    csv_reader=csv.DictReader(file,delimiter=",")
    for row in csv_reader:
        # load expenses
        past_expense={}
        past_expense["name"] = row["NAME"]
        past_expense["date"] = row["DATE"]
        past_expense["category"] = row["CATEGORY"]
        past_expense["amount"] = row["AMOUNT"]
        past_expense["description"] = row["DESCRIPTION"]
        past_expenses.append(past_expense)
        
    
def print_Menu():
    print("***********************************************************************")
    print("###################### PERSONAL EXPENSE TRACKER #######################")
    print("***********************************************************************")
    print()
    print()
    name=input("Please enter your name :")
    print()
    print("Welcome ",name," !!!!!!!!")
    load_expense()
    while(True):
        print()
        print()
        print("***********************************************************************")
        print("############################## MAIN MENU ##############################") 
        print("***********************************************************************")
        print()
        print()
        print("Select your choice as per the given Menu")
        print()
        print("Press 1 to ADD EXPENSE")
        print("Press 2 to VIEW EXPENSE")
        print("Press 3 to TRACK BUDGET")
        print("Press 4 to SAVE EXPENSE")
        print("Press 5 to EXIT")
        print()
        print()
        choice=int(input("Enter your choice : "))
        print()
        print()
        if choice == 1:
            add_expense(name)
            print("Navigating back to main menu")
        elif choice == 2:
            view_expense(name)
            print("Navigating back to main menu")
        elif choice == 3:
            track_budget(name)
            print("Navigating back to main menu")
        elif choice == 4:
            save_expense(name)
            print("Navigating back to main menu")
        elif choice == 5:
            print("Shutting down")
            print()
            print()
            print("***********************************************************************")
            break
        else:
            print("INVALID CHOICE!!!!!!! Please try again")
            print("Navigating back to main menu")

        
print_Menu()
        
```
